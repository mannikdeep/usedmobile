```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
url = 'https://raw.githubusercontent.com/mannikdeep/usedmobile/main/used_device_data.csv'
```

```python
df = pd.read_csv(url)
df
```

```python
df['os'].unique()
```

```python
df['device_brand'].unique()
```

```python
def os_changer(x):
  if x == 'Android':
    return 0
  elif x == 'iOS':
    return 1
  elif x == 'Others':
    return 4
  elif x == 'Windows':
    return 3
```

```python
df['os'] = df['os'].apply(os_changer)
```

```python
df.columns
```

```python
def brand_changer(x):
  if x == 'Samsung':
    return 0
  elif x == 'Realme':
    return 1
  elif x == 'Vivo':
    return 2
  elif x == 'Xiaomi':
    return 3
  elif x == 'Apple':
    return 4
  elif x == 'Google':
    return 5
  elif x == 'Oneplus':
    return 6
  elif x == 'Motorola':
    return 7
  elif x == 'Oppo':
    return 8
  elif x == 'Nokia':
    return 9
  elif x == 'Honor':
    return 10
  else:
    return 11
```

```python
df['device_brand'] = df['device_brand'].apply(brand_changer)
```

```python
df.head()
```

```python
def fiveg(x):
  if x == 'no':
    return 0
  return 1
```

```python
df['5g'] = df['5g'].apply(fiveg)
```

```python
df.head()
```

```python
def price(x):
  return int(x * 100) * 62
```

```python
df['price'] = df['normalized_used_price'].apply(price)
```

```python
df.head()
```

```python
main = df.loc[:,['device_brand','5g','internal_memory','ram','days_used','price']]
```

```python
main = main.dropna()
```

```python
x = main.loc[:,'device_brand':'days_used']
y = main['price']
```
